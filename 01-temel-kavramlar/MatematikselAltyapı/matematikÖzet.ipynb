# Genel Özet  

Bu bölümde kuantum mekaniğini anlamak için gerekli matematiksel temelleri inceledik. Klasik lineer cebir, olasılık ve analiz kavramlarını kuantum dünyasına uyarlayarak soyut matematiğin fiziğe nasıl bağlandığını gördük.  

## Temel Noktalar  

- **Vektörler ve Matrisler:**  
  Kuantum durumları vektörlerle, dönüşümler ve ölçümler matrislerle ifade edilir.  

- **İç Çarpım ve Hilbert Uzayı:**  
  İç çarpım sayesinde vektörlerin uzunluğu ve dikliği tanımlanır. Hilbert uzayı, kuantum durumlarının yaşadığı soyut mekândır.  

- **Karmaşık Sayılar:**  
  Dalga fonksiyonları karmaşık değerli olup faz bilgisi taşır. Ölçüm olasılıkları mutlak değer karesiyle belirlenir.  

- **Olasılık ve İstatistik:**  
  Ölçüm sonuçları deterministik değil olasılıksaldır. Beklenen değer, varyans ve dağılımlar kuantum deneylerin yorumunda kullanılır.  

- **Analiz (Limit, Türev, İntegral, Fourier):**  
  Dalga fonksiyonlarının sürekliliği, Schrödinger denklemi ve konum–momentum ilişkisi analiz kavramlarına dayanır. Fourier dönüşümü belirsizlik ilkesinin matematiksel temelini sağlar.  

- **Operatörler:**  
  Fiziksel gözlemlenebilirler Hermit operatörlerle, zaman evrimi ise üniter operatörlerle temsil edilir. Özdeğer–özvektör yapısı ölçüm postülatının kalbidir.  

- **Soyutlamadan Yoruma:**  
  - Dalga fonksiyonu ölçüm olasılıklarını belirler.  
  - Süperpozisyon ilkesi sayesinde sistemler aynı anda birden çok durumda bulunabilir.  
  - Ölçüm postülatı, ölçümün sonucu kadar sistemin son halini de belirler.  
  - Dolanıklık, klasik dünyada eşi olmayan kuantum korelasyonlarını tanımlar.  

---

## Hatırlanması Gerekenler  

1. Kuantum mekaniği **olasılıksal** bir kuramdır.  
2. Tüm olasılıkların toplamı her zaman 1’dir (**normalizasyon**).  
3. Ölçüm yapıldığında sistem **çöker** ve sonuç özdeğerlerden biri olur.  
4. Çok parçacıklı sistemlerde **tensor çarpımı** kullanılır.  
5. Dolanıklık, kuantum bilgi biliminin temel taşıdır.  

---

## Bölüm Sonu Soruları (Düşün – Çalış – Öğren)  

1. Bir parçacığın dalga fonksiyonu  
   $$
   \psi(x) = \frac{1}{\sqrt{L}}, \quad 0<x<L
   $$  
   olsun. Normalizasyonu ve ortalama konumu $\langle x \rangle$’i hesaplayın.  

2. Spin-$\tfrac{1}{2}$ parçacığı  
   $$
   |\psi\rangle = \tfrac{\sqrt{3}}{2}|0\rangle + \tfrac{1}{2}|1\rangle
   $$  
   durumunda. Spin-$z$ ölçümünde olasılıkları bulun.  

3. Fourier dönüşümü ile Gauss fonksiyonunun yine Gauss olduğunu gösterin. Sonucu belirsizlik ilkesi ile yorumlayın.
