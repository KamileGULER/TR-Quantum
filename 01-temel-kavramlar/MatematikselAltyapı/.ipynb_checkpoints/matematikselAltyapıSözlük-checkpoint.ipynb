# Kuantum Mekaniği Terimler Sözlüğü  

Bu sözlük, *“Kuantumu Anlamak İçin Gerekli Matematiksel Temeller”* başlığı altında işlenen bütün bölümlerde geçen temel kavramları açıklar.  

---

## 1. Giriş  

- **Soyutlama:** Karmaşık fiziksel olayların, matematiksel semboller ve denklemlerle sade bir biçimde ifade edilmesi.  
- **Matematiksel dil:** Fizikte, özellikle kuantum mekaniğinde, olayları ifade etmek için kullanılan evrensel dil.  
- **Modelleme:** Fiziksel gerçekliğin matematiksel yapılar (vektör, fonksiyon, operatör) ile temsil edilmesi.  
- **Postülat:** Bir teorinin temeline yerleştirilen, doğruluğu sorgulanmadan kabul edilen temel ilke. Kuantum mekaniği birkaç postülat üzerine kuruludur.  

---

## 2. Temel Matematiksel Araçlar  

### Lineer Cebir  

- **Vektör:** Birden çok bileşenden oluşan matematiksel nesne; kuantumda bir sistemin durumunu temsil eder.  
- **Matris:** Satır ve sütunlardan oluşan tablo; vektörler üzerinde dönüşüm yapan operatörlerin temsil biçimi.  
- **Lineer dönüşüm:** Vektörleri başka vektörlere lineer olarak dönüştüren işlem.  
- **Özvektör:** Operatör uygulandığında yalnızca boyu değişen, yönü sabit kalan özel vektör.  
- **Özdeğer:** Özvektörün operatör etkisiyle ölçeklendiği sayı. Ölçüm sonuçlarını temsil eder.  
- **Tensor çarpımı:** İki vektörü birleştirerek daha yüksek boyutlu bir vektör oluşturur. Çok parçacıklı sistemlerin tanımı için gereklidir.  

### Karmaşık Sayılar  

- **Karmaşık sayı ($z=a+ib$):** Reel ($a$) ve sanal ($b$) kısımlardan oluşur, $i^2=-1$.  
- **Karmaşık düzlem:** Karmaşık sayıların geometrik gösterimi (Reel eksen, sanal eksen).  
- **Modül ($|z|$):** Karmaşık sayının büyüklüğü, $|z|=\sqrt{a^2+b^2}$.  
- **Faz:** Karmaşık sayının açısı. Kuantumda girişim ve faz farkı fenomenleri için kritik.  
- **Karmaşık eşlenik ($z^*$):** $a+ib$’nin eşleniği $a-ib$. İç çarpım tanımında kullanılır.  

### İç Çarpım ve Hilbert Uzayı  

- **İç çarpım:** İki vektörü alıp skaler üreten işlem. Benzerliği ve ortogonalliği ölçer.  
  $$
  \langle u,v \rangle = \sum_i u_i^* v_i
  $$  
- **Norm:** Bir vektörün uzunluğu. $\|v\| = \sqrt{\langle v,v \rangle}$.  
- **Ortogonallik:** İki vektörün iç çarpımının sıfır olması (dik olmaları).  
- **Hilbert uzayı:** İç çarpımlı ve tam bir vektör uzayı; kuantum durumlarının matematiksel evreni.  
- **Ortonormal baz:** Bir uzayı oluşturan, birbirine dik ve normu 1 olan vektörler kümesi.  

### Olasılık ve İstatistik  

- **Olasılık:** Bir olayın gerçekleşme derecesi ($0 \leq P \leq 1$).  
- **Olasılık dağılımı:** Rastgele değişkenin alabileceği değerler ve olasılıklarının tamamı.  
- **Beklenen değer (ortalama):** Rastgele değişkenin uzun vadede ortalama sonucu.  
  $$
  \mathbb{E}[X] = \sum_i x_i p_i \quad \text{veya} \quad \int x f(x)\,dx
  $$  
- **Varyans:** Ölçüm sonuçlarının ortalama etrafındaki yayılımı.  
- **Standart sapma:** Varyansın karekökü; belirsizlik ölçüsü.  
- **Normalizasyon:** Tüm olasılıkların toplamı 1 olması gereği.  

---

## 3. Analizden Gerekli Kavramlar  

- **Limit:** Fonksiyonun belirli bir noktaya yaklaşırken aldığı değer.  
- **Süreklilik:** Fonksiyonun grafiğinde “kopma” olmaması. Dalga fonksiyonları fiziksel olarak genellikle süreklidir.  
- **Türev:** Fonksiyonun değişim hızı. Schrödinger denklemi ikinci türev içerir.  
- **İntegral:** Fonksiyonun toplam etkisini veya alanını hesaplar. Normalizasyon ve beklenen değerler integralle bulunur.  
- **Fourier dönüşümü:** Bir fonksiyonun farklı bir bazda (ör. konum yerine momentum) ifade edilmesini sağlar.  
- **Belirsizlik ilkesi:** Konum ve momentum belirsizliklerinin çarpımı hiçbir zaman $\hbar/2$’den küçük olamaz.  

---

## 4. Operatörler ve Lineer Cebir  

- **Operatör:** Hilbert uzayında tanımlı fonksiyon; kuantumda fiziksel büyüklükleri temsil eder.  
- **Hermit operatör:** Özdeğerleri daima gerçek olan, ölçülebilir nicelikleri temsil eden operatör.  
- **Üniter operatör:** Normu koruyan dönüşüm. Kuantumda zaman evrimini tanımlar.  
- **Özdeğer problemi:** $A|a\rangle = a|a\rangle$ eşitliği. Ölçüm sonuçları $a$ değerleridir.  
- **Komütatör:** $[A,B]=AB-BA$. Eğer sıfırsa iki büyüklük aynı anda ölçülebilir.  
- **Hamiltonyen ($H$):** Enerji operatörü; sistemin zaman evrimini belirler.  

---

## 5. Soyutlamadan Fiziksel Yoruma  

- **Dalga fonksiyonu ($\psi$):** Bir kuantum sisteminin tam durumunu tanımlayan karmaşık fonksiyon. Ölçüm olasılıkları $|\psi|^2$ ile verilir.  
- **Olasılık genliği:** Dalga fonksiyonunun kendisi. Gözlemlenemez, ancak karesinin mutlak değeri ölçüm olasılıklarını belirler.  
- **Normalizasyon:** Dalga fonksiyonunun olasılık yoğunluğunun tüm uzayda 1’e eşit olması:  
  $$
  \int |\psi(x)|^2 dx = 1
  $$  
- **Süperpozisyon:** Kuantum sistemlerinin aynı anda birden fazla durumda bulunabilmesi. Ölçüm yapıldığında sonuçlardan biri olasılıksal olarak elde edilir.  
- **Postülat:** Teorinin temelini oluşturan, doğruluğu sorgulanmadan kabul edilen ilke.  
- **Ölçüm postülatı:** Ölçüm sonucu, gözlemlenebilirin özdeğerlerinden biridir; ölçümden sonra sistem ilgili özvektöre çöker.  
- **Çökme (collapse):** Ölçüm sonrası dalga fonksiyonunun süperpozisyondan tek bir özvektöre dönüşmesi.  
- **Hermit operatör:** Kuantum ölçümlerini temsil eden operatör. Özdeğerleri ölçüm sonuçlarıdır.  
- **Özvektör / Özdeğer:** Ölçüm operatörünün özel çözümleri. Özdeğer → ölçüm sonucu, Özvektör → ölçüm sonrası durum.  
- **Dolanıklık (entanglement):** Çok parçacıklı sistemlerde birleşik durumun tekil sistemlerin çarpımına ayrılamaması.  
- **Bell durumu:** En temel dolanık durum. Örn:  
  $$
  |\Phi^+\rangle = \tfrac{1}{\sqrt{2}} (|00\rangle + |11\rangle)
  $$  
- **Korelasyon:** İki parçacığın ölçüm sonuçları arasındaki bağ. Dolanıklık, klasik korelasyondan daha güçlüdür.  
- **Heisenberg belirsizlik ilkesi:** Konum ve momentum ölçümlerinin belirsizlikleri çarpımı asla $\hbar/2$’den küçük olamaz.  

---

## 6. Sonuç  

- **Matematik = Dil:** Kuantum mekaniğini anlamak, aslında onun matematiksel dilini öğrenmektir.  
- **Temel taşlar:** Lineer cebir, karmaşık sayılar, Hilbert uzayı, olasılık, analiz, operatörler.  
- **Uygulama:** Bu temeller, entanglement, kuantum bilgi ve kuantum bilgisayar gibi ileri konuların anlaşılması için zemin oluşturur.
