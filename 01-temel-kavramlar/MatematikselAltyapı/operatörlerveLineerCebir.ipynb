# Operatörler

Kuantum mekaniğinde operatörler, **fiziksel büyüklüklerin matematiksel karşılığı**dır. Bir büyüklüğü ölçmek, aslında o büyüklüğün operatörünü sistemin dalga fonksiyonuna uygulamaktır.

---

## Temel Tanım

Bir operatör, Hilbert uzayında tanımlı bir fonksiyondur:

$$
\hat{A} : \mathcal{H} \to \mathcal{H}, \quad \hat{A}|\psi\rangle = |\phi\rangle
$$

Burada:
- $|\psi\rangle$: sistemin başlangıç durumu  
- $|\phi\rangle$: operatör uygulandıktan sonraki durum  

---

## Hermit (Öz-adjoint) Operatörler

- Ölçülebilir nicelikler Hermit operatörlerle ifade edilir.
- Tanım:

$$
\langle \phi | \hat{A}\psi \rangle = \langle \hat{A}\phi | \psi \rangle
$$

- Özellikleri:
  - Özdeğerleri her zaman **gerçek sayıdır**.
  - Farklı özdeğerlere karşılık gelen özvektörler **ortogonaldir**.

**Örnek:**  
Konum operatörü

$$
\hat{x}\psi(x) = x \psi(x)
$$

Momentum operatörü

$$
\hat{p}\psi(x) = -i\hbar \frac{d}{dx}\psi(x)
$$

---

## Üniter Operatörler

- Sistemlerin zamanla nasıl değiştiğini tanımlar.  
- Tanım:

$$
\hat{U}^\dagger \hat{U} = \hat{U}\hat{U}^\dagger = I
$$

- Özellikleri:
  - Normu korur: $\|\hat{U}|\psi\rangle\| = \||\psi\rangle\|$
  - Tersleri vardır: $\hat{U}^{-1} = \hat{U}^\dagger$

**Örnek:** Zaman evrimi operatörü  

$$
\hat{U}(t) = e^{-i\hat{H}t/\hbar}
$$

---

## Projeksiyon Operatörleri

- Bir duruma projeksiyon yapar:

$$
\hat{P} = |a\rangle \langle a|
$$

Özellikleri:
- İdempotenttir: $\hat{P}^2 = \hat{P}$
- Hermitiktir: $\hat{P}^\dagger = \hat{P}$

**Fiziksel anlam:** Bir ölçüm sonucunu elde etme olasılığını verir.

---

## Komütatör ve Belirsizlik

İki operatörün **komütatörü**:

$$
[\hat{A}, \hat{B}] = \hat{A}\hat{B} - \hat{B}\hat{A}
$$

- Eğer $[\hat{A},\hat{B}]=0$ ise aynı anda ölçülebilirler.  
- Eğer $[\hat{A},\hat{B}]\neq 0$ ise belirsizlik vardır.  

**Örnek:**

$$
[\hat{x},\hat{p}] = i\hbar
$$

Bu ifade Heisenberg belirsizlik ilkesine yol açar.

---

## Özdeğer–Özvektör İlişkisi

Her operatör için:

$$
\hat{A}|a\rangle = a|a\rangle
$$

- $a$: özdeğer (ölçüm sonucu)  
- $|a\rangle$: özvektör (sistemin olası durumu)  

**Örnek (Spin-$z$):**

$$
\sigma_z = \begin{bmatrix}1 & 0 \\ 0 & -1\end{bmatrix}
$$

$$
\sigma_z |0\rangle = +1 |0\rangle, \quad
\sigma_z |1\rangle = -1 |1\rangle
$$

---

## Operatörlerin Fiziksel Rolleri

1. **Ölçüm:** Fiziksel nicelikleri ölçmek → Hermit operatörler  
2. **Dinamik:** Sistemin zaman evrimini belirlemek → Üniter operatörler  
3. **Projeksiyon:** Ölçüm sonrası sistemin çöktüğü durumu ifade etmek  
4. **Belirsizlik:** Komütatörler aracılığıyla aynı anda ölçülebilirlik sınırlarını göstermek  

---

## Özet

- **Operatörler**, kuantum mekaniğinde fiziksel büyüklüklerin dilidir.  
- **Hermit operatörler**, ölçülebilir nicelikleri temsil eder ve sonuçları her zaman gerçektir.  
- **Üniter operatörler**, kuantum sisteminin zamanla nasıl evrildiğini gösterir ve normu korur.  
- **Projeksiyon operatörleri**, ölçüm sonrası sistemin hangi duruma çökeceğini belirler.  
- **Komütatörler**, aynı anda ölçülebilirlik koşullarını ortaya koyar ve belirsizlik ilkesinin matematiksel temelini sağlar.  

Kısacası: Operatörler olmadan kuantum mekaniğinde ne ölçüm yapabiliriz ne de sistemin zamanla nasıl değiştiğini anlayabiliriz.
