# Kuantum Kapıları ve Devreler

## Giriş

### Hatırlatma

Kuantum bilgi işlemeye başlamadan önce, temel yapı taşlarını tekrar hatırlamak faydalı olacaktır.

- **Qubit**: Klasik bir bit yalnızca $0$ veya $1$ durumunda olabilirken, bir qubit bu iki durumun süperpozisyonunda bulunabilir:

$$
|\psi\rangle = \alpha |0\rangle + \beta |1\rangle, \quad \alpha, \beta \in \mathbb{C}, \; |\alpha|^2 + |\beta|^2 = 1
$$

- **Süperpozisyon**: Yukarıdaki denklemde görüldüğü gibi, bir qubit aynı anda hem $|0\rangle$ hem de $|1\rangle$ durumlarının lineer kombinasyonunda bulunur.

- **Dolanıklık (Entanglement)**: İki veya daha fazla qubitin ortak bir kuantum durumunda bulunmasıdır. Örneğin Bell durumu:

$$
|\Phi^+\rangle = \frac{1}{\sqrt{2}} \Big(|00\rangle + |11\rangle \Big)
$$

Klasik sistemlerde mümkün olmayan bu tür korelasyonlar, kuantum hesaplamanın gücünü ortaya çıkarır.

---

### Kuantum Kapıları

Kuantum algoritmalarının temeli, qubitler üzerinde gerçekleştirilen **kuantum kapılarıdır**.  
Klasik bilgisayarlarda lojik kapılar (AND, OR, NOT) bilgi işleme birimiyken, kuantum bilgisayarlarda tüm işlemler **üniter matrislerle** tanımlanır.

Bir qubit üzerindeki işlem, Bloch küresi üzerinde bir dönüşüm (rotasyon) ya da faz kaydırma olarak düşünülebilir.  
Genel olarak bir kuantum kapısı:

$$
U \in \mathbb{C}^{2^n \times 2^n}, \quad U^\dagger U = I
$$

şeklinde tanımlanan bir **üniter dönüşümdür**.

---

### Tek ve Çok Qubit Kapıları

- **Tek qubit kapıları**: Bir qubit üzerinde dönüşüm veya faz kaydırma yapar (örneğin $X, Y, Z, H, S, T, R_x, R_y, R_z$).

- **İki veya daha fazla qubit kapıları**: Kontrol mekanizmaları veya dolanıklık üretimi için kullanılır (örneğin $CNOT, CZ, SWAP, Toffoli$).

Bu kapıların birleşimi ile **kuantum devreleri** oluşturulur.  
Kuantum devreleri, bir algoritmanın qubitler üzerinde adım adım nasıl çalıştığını görselleştiren ve uygulayan temel yapıdır.

---

### Bu Bölümde Öğrenecekleriniz

- Qubitler üzerinde işlem yapan **temel tek qubit kapıları** ($X, Y, Z, H, S, T, R_x, R_y, R_z$)  
- İki ve daha fazla qubit ile çalışan **kontrollü kapılar** ($CNOT, CZ, SWAP, Toffoli$)  
- Kapıların **matematiksel gösterimleri** ve **Bloch küresi etkileri**  
- Kapıların birleşimiyle oluşan **kuantum devreleri**  
- Qiskit kullanarak kapıların ve devrelerin **simülasyonlarını çalıştırma**
- 
---

Bu bölümde, kuantum bilgi işlemenin temel yapı taşları olan kapılar ve devrelerle tanışacaksınız.  
İlk olarak tek qubit kapılarının matematiksel ve geometrik yorumlarını öğrenecek, ardından çok qubit kapılarıyla dolanıklık üretimini keşfedeceksiniz.  
Son aşamada bu kapıların nasıl birleşerek kuantum devrelerini oluşturduğunu ve algoritmalara temel hazırladığını göreceksiniz.  

Kuantum kapıları ve devreleri, kuantum algoritmalarının temelini oluşturur.  
Bu konuyu öğrenmek, **dolanıklık üretme, süperpozisyon kullanma ve klasik hesaplamada çözülemeyen problemleri etkin şekilde çözme** yolunda atılan en önemli adımdır.
