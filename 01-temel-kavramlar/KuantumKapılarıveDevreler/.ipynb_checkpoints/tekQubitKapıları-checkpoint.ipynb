# Tek Qubit Kapıları

Tek qubit kapıları, $2\times 2$ boyutlu **üniter matrislerle** tanımlanır.  
(Üniter matris demek, $U^\dagger U = I$ eşitliğini sağlayan, yani tersinir ve normu koruyan matris demektir.  
Qubitler üzerinde uygulandığında olasılıkların korunmasını sağlar.)  

---

## Pauli Kapıları

### X Kapısı (Pauli-X, NOT Kapısı)

$$
X=\begin{bmatrix}0&1\\1&0\end{bmatrix}
$$

(Bu matris $|0\rangle \to |1\rangle$ ve $|1\rangle \to |0\rangle$ dönüşümü yapar.  
Yani klasik NOT kapısının kuantum karşılığıdır. Bloch küresi üzerinde x-ekseni etrafında $\pi$ kadar döndürür.)

Etkiler:
- $(X|0\rangle = |1\rangle)$
- $(X|1\rangle = |0\rangle)$


```python
from qiskit import * 
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline
```

```python
qc = QuantumCircuit(1)
qc.x(0)
```

```python
_ = qc.draw(output='mpl')  
```

---
### Y Kapısı (Pauli-Y)

$$
Y=\begin{bmatrix}0&-i\\i&0\end{bmatrix}
$$

(Burada $i$, kompleks sayıların hayali birimidir. Bu matris hem bit çevirme hem de faz ekleme yapar.  
$|0\rangle$’i $i|1\rangle$’e, $|1\rangle$’i $-i|0\rangle$’e dönüştürür. Bloch küresi üzerinde y-ekseni etrafında $\pi$ kadar döndürür.)

Etkiler:
- $(Y|0\rangle = i|1\rangle)$  
- $(Y|1\rangle = -i|0\rangle)$

```python
qc = QuantumCircuit(1)
qc.y(0)
```

```python
_ = qc.draw(output='mpl')  
```

---
### Z Kapısı (Pauli-Z)

$$
Z=\begin{bmatrix}1&0\\0&-1\end{bmatrix}
$$

(Bu matris $|0\rangle$’i değiştirmez ama $|1\rangle$ durumunun fazını ters çevirir: $|1\rangle \to -|1\rangle$.  
Bloch küresi üzerinde z-ekseni etrafında $\pi$ rotasyonu yapar.)

Etkiler:
- $(Z|0\rangle = |0\rangle)$  
- $(Z|1\rangle = -|1\rangle)$
  

```python
qc = QuantumCircuit(1)
qc.z(0)
```

```python
_ = qc.draw(output='mpl')  
```

---
## Hadamard (H) Kapısı

$$
H=\tfrac{1}{\sqrt{2}}\begin{bmatrix}1&1\\1&-1\end{bmatrix}
$$

(Bu kapı $|0\rangle$ durumunu $\tfrac{|0\rangle+|1\rangle}{\sqrt{2}}$ haline getirir.  
Yani qubit’i süperpozisyona sokar. $|0\rangle$ ve $|1\rangle$’in eşit ihtimalli bir karışımını oluşturur.)

Etkiler:
- $(H|0\rangle = \tfrac{1}{\sqrt{2}}(|0\rangle + |1\rangle))$  
- $(H|1\rangle = \tfrac{1}{\sqrt{2}}(|0\rangle - |1\rangle))$

Yani tek bir qubit’i **süperpozisyon** durumuna sokar.


```python
qc = QuantumCircuit(1)
qc.h(0)
```

```python
_ = qc.draw(output='mpl')  
```

## Faz Kapıları (S ve T)

Faz kapıları, qubit’in **genliğini değiştirmeden yalnızca fazını kaydıran** özel kapılardır.  
(Yani ölçüm olasılıklarını değiştirmezler; $|0\rangle$ ve $|1\rangle$ durumlarının ihtimalleri aynı kalır.  
Fakat qubit’in **göreli fazı** değiştiği için girişim deneylerinde veya başka kapılarla birleştiğinde ölçüm sonuçları üzerinde etkili olurlar.)  

Başka bir deyişle faz kapıları, kuantum bilgiyi **faz açısı** üzerinden kodlamak için kullanılır.  
Bu nedenle kuantum algoritmalarında kritik öneme sahiptirler.  

---

### S Kapısı

$$
S=\begin{bmatrix}1&0\\0&i\end{bmatrix}
$$

(Bu kapı $|0\rangle$’i değiştirmez, $|1\rangle$ durumunun fazını $90^\circ$ döndürür.  
Yani $|1\rangle \to i|1\rangle$ dönüşümü olur. $S^2 = Z$ eşitliği geçerlidir.  
Başka kapılarla birlikte kullanıldığında, $Z$ ekseni etrafındaki $\pi/2$’lik bir rotasyona denk gelir.)  

```python
qc = QuantumCircuit(1)
state0 = Statevector.from_instruction(qc)
plot_bloch_multivector(state0)
plt.show()
```

Başlangıçta qubit $|0\rangle$ durumundadır.  
Bloch küresinde bu, **kuzey kutbunda z-ekseni yönünde** duran bir vektör olarak görünür.  
Bu noktada ölçüm her zaman $|0\rangle$ sonucunu verir.

```python
qc.h(0) #başlangıç durumu |0> -> |+>
```

```python
_ = qc.draw(output='mpl')  
```

```python
# Bloch küresinde başlangıç durumunu çiz
state1 = Statevector.from_instruction(qc)
plot_bloch_multivector(state1)
plt.show()
```

Qubit $|0\rangle$ durumundan Hadamard kapısıyla süperpozisyona geçti.  
Vektör x-ekseni yönünü gösteriyor, bu da $|+\rangle=\tfrac{|0\rangle+|1\rangle}{\sqrt{2}}$ durumuna karşılık geliyor.

```python
# S kapısını uygula
qc.s(0)
```

```python
# Yeni durumu al
state2 = Statevector.from_instruction(qc)
plot_bloch_multivector(state2)
plt.show()
```

S kapısı $|1\rangle$ bileşenine faz ekledi ve vektör x-ekseni yerine y-ekseni yönüne döndü.  
Bu yeni yön $|+i\rangle=\tfrac{|0\rangle+i|1\rangle}{\sqrt{2}}$ durumunu gösteriyor; olasılıklar aynı kaldı ama faz değişti.

### T Kapısı

$$
T=\begin{bmatrix}1&0\\0&e^{i\pi/4}\end{bmatrix}
$$

(Bu kapı $|0\rangle$ durumunu sabit bırakır, $|1\rangle$ durumunun fazını $45^\circ$ kaydırır.  
Yani $|1\rangle \to e^{i\pi/4}|1\rangle$. Burada $e^{i\pi/4}=\cos(\pi/4)+i\sin(\pi/4)=\tfrac{\sqrt{2}}{2}+i\tfrac{\sqrt{2}}{2}$,  
yani karmaşık düzlemde birim çember üzerinde $45^\circ$’lik bir dönüşümdür.)  

#### Özellikleri  
- **$T^2=S$**: T kapısı iki kez uygulandığında S kapısı elde edilir.  
- **$T^4=Z$**: Dört kez uygulandığında Pauli-Z kapısına eşdeğerdir.  
- **$T^8=I$**: Sekiz kez uygulanırsa birim operatöre geri döner.  
- Bu nedenle T, bazen **Z kapısının sekizinci kökü** (eighth root of Z) olarak da adlandırılır.  

#### Neden Önemli?  
- Clifford kapıları (H, S, CNOT) tek başına **evrensel değildir**; yani tek başlarına tüm kuantum dönüşümlerini üretemezler.  
- T kapısı, Clifford kapı kümesine eklendiğinde **evrensel kapı seti** elde edilir.  
  Bu, **herhangi bir kuantum algoritmasının** bu kapıların birleşimiyle gerçekleştirilebileceği anlamına gelir.  
- Kuantum hata düzeltme (quantum error correction) ve derleme (compilation) çalışmalarında “T-gate count” yani T kapısı sayısı,  
  kuantum algoritmalarının pratik maliyetini belirleyen en önemli metriklerden biridir.  

#### Bloch Küresi Yorumu  
T kapısı, Bloch küresi üzerinde **Z ekseni etrafında $\pi/4$’lük bir rotasyon** yapar.  
Yani qubit’in olasılıklarını değiştirmez ama fazını kaydırarak girişim desenlerini etkiler.
