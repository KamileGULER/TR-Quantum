### Aer / AerSimulator = 
Kuantum devrelerini gerçek cihaz olmadan bilgisayarda çalıştırmaya yarayan simülatör.
### Arama Problemi = 
Büyük bir listede belirli bir girdiyi bulma sorunu.
### Asal Çarpanlara Ayırma =
Bir sayıyı asal sayıların çarpımı şeklinde yazma işlemi; Shor’un çözdüğü problem.
### Ayrık Fourier Dönüşümü (DFT) = 
Verilerin frekans bileşenlerine ayrılmasını sağlayan matematiksel dönüşüm.
### Beauregard Devresi = 
Shor algoritmasının optimize edilmiş kuantum devre tasarımı.
### Beklenen Değer (Ortalama) = 
Bir rastgele değişkenin uzun vadede alacağı ortalama değer.
### Bell Durumu (EPR Çifti) = 
İki qubit’in en basit ve en bilinen dolanıklık hali; |00⟩ ve |11⟩ süperpozisyonu.
### Belirsizlik İlkesi (Heisenberg) = 
Bir parçacığın konumu ve momentumu gibi büyüklüklerin aynı anda kesin olarak ölçülemeyeceğini söyleyen kuantum kuralı.
### Bloch Küresi = 
Qubit’in durumunu 3 boyutlu bir küre üzerinde görselleştirme yöntemi.
### Boole Fonksiyonu = 
Girdi olarak 0 ve 1 alan ve çıktı olarak 0 ya da 1 döndüren fonksiyon.
### CHSH Testi = 
Dolanıklığın varlığını ve klasik fiziğe aykırılığını test eden deneysel yöntem.
### Circuit (Devre) = 
Qubit’ler üzerinde yapılan işlemlerin adım adım gösterildiği yapı.
### ClassicalRegister = 
Qiskit’te klasik (0 ve 1) bilgilerin tutulduğu kayıt yapısı.
### Clbit = 
Qiskit’te tek bir klasik bit nesnesi.
### Clifford Grubu = 
Pauli ve Hadamard gibi kapıları içeren özel kapı ailesi.
### CNOT Kapısı (Kontrollü-NOT) = 
Bir qubit’in durumuna bağlı olarak diğerini çeviren kapı. Dolanıklık oluşturmakta kullanılır.
### Controlled-X (CNOT) Kapısı = 
Kontrol qubit 1 olduğunda hedef qubit’i çeviren kapı. (Dolanıklık üretmekte kullanılır.)
### Controlled-Z (CZ) Kapısı = 
Kontrol qubit 1 olduğunda hedef qubit’in fazını değiştiren kapı.
### CZ (Kontrollü-Z) = 
Kontrol qubit’i 1 olduğunda hedef qubit’in fazını değiştiren kapı.
### ÇOK Qubit Kapıları = 
İki veya daha fazla qubit üzerinde aynı anda işlem yapan kapılar.
### Dalga Fonksiyonu = 
Bir kuantum sisteminin olasılık dağılımını belirleyen matematiksel ifade.
### Dalga Paketi = 
Bir parçacığın konum ve momentum bilgilerini birlikte tanımlayan dalga fonksiyonu kombinasyonu.
### Delta Potansiyeli = 
Çok dar bir bölgede sonsuz büyüklükte olan, idealize edilmiş potansiyel modeli.
### Dengeli Fonksiyon = 
Çıktılarının yarısı 0, yarısı 1 olan fonksiyon.
### Dengeli Süperpozisyon = 
Qubit’in %50 olasılıkla 0, %50 olasılıkla 1 çıkacak şekilde hazırlanması.
### Derleyiciler (Compiler) = 
Kuantum devresini, seçilen donanımın çalıştırabileceği kapılara çeviren araçlar.
### Deutsch-Jozsa Algoritması = 
Fonksiyonun sabit mi yoksa dengeli mi olduğunu tek adımda belirleyen kuantum algoritması.
### Deutsch-Jozsa Problemi = 
Verilen bir Boole fonksiyonunun sabit mi yoksa dengeli mi olduğunu belirleme problemi.
### Devre Diyagramı = 
Kuantum devresindeki kapıların qubit’lere nasıl uygulandığını gösteren şema.
### Difüzyon Operatörü (Grover Diffusion Operator) = 
Olasılık genliklerini tersine çevirerek doğru çözümün olasılığını artıran işlem.
### Doğruluk Tablosu = 
Kapıların giriş ve çıkışlarını gösteren tablo.
### E-bit = 
Bir dolanık qubit çifti; kuantum iletişiminde temel kaynak.
### Estimation (Tahmin Algoritmaları) = 
Özellikle faz tahmini gibi kuantum algoritmalarında kullanılan yöntemler.
### Euler Açılımı = 
Bir qubit üzerindeki dönüşümlerin üç temel rotasyonla ifade edilmesi yöntemi.
### Faz = 
Karmaşık sayının açısal yönünü belirleyen büyüklük.
### Faz Geri Tepmesi (Phase Kickback) = 
Kontrol qubit’inin hedef operatörün faz bilgisini üzerine “geri yansıtması” olayı.
### Fourier Dönüşümü = 
Bir fonksiyonun frekans bileşenlerine ayrılmasını sağlayan matematiksel işlem.
### Fredkin (CSWAP) Kapısı = 
Kontrol qubit’i 1 olduğunda iki qubit’in yerini değiştiren kapı.
### Gauss Fonksiyonu = 
Çan eğrisi şeklindeki özel matematiksel fonksiyon; kuantumda dalga paketlerini tanımlamakta kullanılır.
### Grover Algoritması = 
Arama problemlerini klasik yöntemlerden daha hızlı çözen kuantum algoritması.
### Hadamard Kapısı = 
Qubit’i süperpozisyona sokan kapı.
### Hadamard Kapısı = 
Qubit’i süperpozisyona sokar. 0 ve 1’i dengeli şekilde karıştırır.
### Hata Düzeltme = 
Çok qubit kapılarının, kuantum hesaplamalarda hataları gidermek için kullanımı.
### Hedef Qubit = 
Çoklu qubit kapılarında, kontrolün etki ettiği qubit.
### Hermit Operatör = 
Kendi eşlenik transpozuna eşit olan operatör; kuantumda gözlemlenebilir nicelikleri temsil eder.
### Hilbert Uzayı = 
Qubit’lerin matematiksel olarak tanımlandığı vektör uzayı.
### Histogram = 
Ölçüm sonuçlarını çubuk grafikle gösterme yöntemi.
### Holevo Sınırı = 
Bir kuantum sisteminin taşıyabileceği klasik bilgi miktarını sınırlayan ilke.
### Işınlanma = 
Günlük dildeki "maddenin taşınması" değil, qubit bilgisinin aktarılmasıdır.
### Initialize = 
Qubit’i istenen başlangıç durumuna getirmek için kullanılan komut.
### İç Çarpım (Skaler Çarpım) = 
İki vektörün benzerliğini ölçen matematiksel işlem. Kuantumda olasılık hesaplamalarında kullanılır.
### İntegral = 
Fonksiyonun altında kalan alanı ya da toplam etkisini ölçen matematiksel işlem.
### İstatistik = 
Olayların uzun vadede hangi sıklıkta ortaya çıktığını inceleyen yöntemler bütünü.
### Kareköklü Hızlanma = 
Grover algoritmasının, klasik aramaya göre yaklaşık √N kat hızlanma sağlaması.
### Karmaşık Düzlem = 
Karmaşık sayıların geometrik olarak gösterildiği düzlem; yatay eksen reel, dikey eksen sanal kısımdır.
### Karmaşık Sayı = 
Reel (gerçek) ve sanal kısımdan oluşan sayı; z = a + ib
### Klasik (Classical) = 
Normal bilgisayarlardaki 0 ve 1.
### Klasik Korelasyon = 
Normal (klasik) sistemlerde ölçüm sonuçları arasındaki önceden belirlenmiş ilişki.
### Klasik Mantık Benzerliği = 
Çok qubit kapılarının AND, OR, NAND gibi klasik kapıların kuantum karşılığı gibi çalışması.
### Klasik Mantık Kapıları (AND, OR, NOT) = 
Klasik bilgisayarlarda kullanılan temel işlemler. Kuantum kapılarına benzetilerek anlatılır.
### Klonlama (No-Cloning) = 
Kuantum bilgisinde bilinmeyen bir qubit’in kopyalanamayacağını söyleyen ilke.
### Komütatör = 
İki operatörün sırasının değişip değişmediğini gösteren işlem.
### Kontrol Qubit = 
Çoklu qubit kapılarında, diğer qubit’in davranışını belirleyen qubit.
### Kontrollü Rotasyon Kapıları (Controlled-R) = 
Küçük açı farkları ekleyerek Fourier dönüşümünü gerçekleştiren kapılar ailesi.
### Kontrol-Hedef İlişkisi = 
Bir qubit’in kontrol, diğerinin hedef olduğu kapı yapısı.
### Konum Operatörü = 
Parçacığın uzaydaki konumunu ölçmeyi temsil eden operatör.
### Konum–Momentum Uzayı = 
Parçacığın hem konum hem de momentum bilgilerini ifade eden matematiksel alan.
### Korelasyonlar = 
Ölçüm sonuçları arasındaki ilişkiler. Kuantumda bu ilişkiler klasik olandan farklıdır.
### Kuantum Devreleri = 
Qubit’ler üzerinde birden fazla kapının sıralı şekilde uygulanmasıyla oluşan yapı.
### Kuantum Dolanıklık (Quantum Entanglement) = 
İki ya da daha çok qubit’in durumlarının birbirine bağlı olması; biri ölçülünce diğeri de anında belirlenir.
### Kuantum Faz Tahmini (QPE) = 
Bir kuantum operatörünün özdeğerindeki fazı ölçmeye yarayan temel algoritma.
### Kuantum Işınlama (Teleportation) = 
Dolanıklık (e-bit) ve klasik bilgi kullanarak bir qubit’in durumunu uzaktan aktarma yöntemi.
### Kuantum Kapıları = 
Qubit’lerin durumunu değiştiren temel işlemler; kuantum algoritmalarının yapı taşları.
### Kuantum Paralelliği = 
Qubit’lerin aynı anda birçok olasılığı işleyebilmesi.
### Limit = 
Bir fonksiyonun değerinin belirli bir noktaya yaklaşırken aldığı değeri tanımlayan kavram.
### Lineer Cebir = 
Vektörler ve matrislerle ilgilenen matematik dalı; kuantum mekaniğinin dili.
### Lineer Dönüşüm = 
Vektörleri başka vektörlere dönüştüren matematiksel işlem.
### Matris = 
Sayıların tablo şeklinde düzenlenmiş hali; kuantumda dönüşümleri temsil eder.
### Measure / Measures = 
Qubit’i ölçme işlemini başlatan komut(lar).
### Modül (Büyüklük) = 
Karmaşık sayının uzunluğu.
### Modüler Üstel Fonksiyon = 
f(x)=a^x modN şeklindeki fonksiyon; Shor algoritmasında kullanılır.
### Momentum Operatörü = 
Parçacığın momentumunu (hareket miktarını) ölçmeyi temsil eden operatör.
### Motivasyon = 
Matematiksel kavramların kuantum mekaniğinde neden gerekli olduğunu açıklayan bağlam.
### Norm = 
Bir vektörün büyüklüğünü ölçen değer.
### Normalizasyon = 
Dalga fonksiyonunun toplam olasılığı 1 olacak şekilde ayarlanması.
### Olasılık = 
Bir olayın gerçekleşme ihtimalini 0 ile 1 arasında bir değerle ifade eden kavram.
### Olasılık Dağılımları = 
Bir rastgele değişkenin hangi değerleri hangi olasılıkla alacağını gösteren dağılım.
### Olasılık Genliği = 
Bir kuantum durumunun ölçüm sonucuna katkısını belirleyen karmaşık sayı. Ölçüm olasılığı genliğin karesinden elde edilir.
### Operatör = 
Kuantum sistemlerinde durumlar üzerinde etki eden matematiksel işlem.
### Operator / Operatör = 
Kuantum devresinde qubit’in durumunu değiştiren matematiksel işlem ya da kapı.
### Oracle (Arama için) = 
Doğru çözümü işaret eden kara kutu fonksiyon.
### Oracle (Kara Kutu Fonksiyon) = 
Bir algoritmada, girdiye karşılık çıktıyı veren ve iç yapısı bilinmeyen fonksiyon.
### Ortogonallik = 
İki vektörün dik olması; kuantumda iki durumun birbirinden bağımsız olduğunu gösterir.
### Ölçüm = 
Qubit’in süperpozisyondan çıkıp ya 0 ya da 1 olması.
### Ölçüm Postülatı = 
Kuantum mekaniğinde, ölçüm yapıldığında sistemin olasılık genliklerine göre belirli bir duruma çökeceğini söyleyen temel kural.
### Özdeğer = 
Bir matrisin belirli bir özvektörüyle ilişkili sabit katsayı.
### Özvektör = 
Bir matrisle çarpıldığında sadece yönü değişmeyen, sabit katsayı ile ölçeklenen vektör.
### Pauli Kapıları = 
Kuantum mekaniğinde temel dönüşümleri yapan kapılar.
### Pauli-X (X Kapısı / NOT) = 
Qubit’i 0’dan 1’e, 1’den 0’a çevirir.
### Pauli-Y (Y Kapısı) = 
Qubit’i Bloch küresinde y-ekseni etrafında döndürür.
### Pauli-Z (Z Kapısı) = 
Qubit’in fazını değiştirir (işaretini ters çevirir).
### Parseval Teoremi = 
Bir fonksiyonun enerjisinin zaman ve frekans uzayında eşit olduğunu söyleyen matematiksel ilke.
### Periyod Bulma = 
Shor algoritmasının temel adımı; belirli bir fonksiyonun periyodunu (tekrar uzunluğunu) bulmak.
### Phi (|Φ⟩) = 
Bell durumları gibi özel dolanıklık durumlarını göstermek için kullanılan sembol.
### Phase Kapısı = 
Qubit’in fazını belirli bir açıyla kaydırır.
### Postülat = 
Kuantum mekaniğinin temel varsayımı ya da kabulü.
### Potansiyel Bariyer / Potansiyel = 
Bir parçacığın hareketini etkileyen enerji engeli ya da enerji fonksiyonu.
### Projeksiyon Operatörü = 
Bir vektörü belirli bir alt uzaya yansıtan operatör.
### Protocol (Protokol) = 
Kuantum ışınlama gibi belirli bir görevi adım adım yerine getiren kurallar bütünü.
### Protokol = 
Belirli bir iletişim veya işlem adımlarının tanımlanmış kuralları.
### Psi (|ψ⟩) = 
Qubit’in veya sistemin genel kuantum durumunu temsil eden sembol.
### QAOA (Quantum Approximate Optimization Algorithm) = 
Zor optimizasyon problemlerini çözmek için kullanılan kuantum algoritması.
### QFT (Quantum Fourier Transform) = 
Kuantum bilgisayarda Fourier dönüşümü.
### Qiskit = 
Kuantum bilgisayarları programlamak için kullanılan yazılım.
### QKD (Quantum Key Distribution) = 
Kuantum dolanıklık kullanılarak güvenli anahtar paylaşımı yapılmasını sağlayan yöntem.
### Quantum Circuit (Kuantum Devresi) = 
Qubit’ler üzerinde işlemlerin sırasıyla tanımlandığı yapı.
### Quantum Register = 
Qiskit’te qubit’lerin tutulduğu kayıt yapısı.
### Qubit = 
Kuantum bilgisayarın en küçük bilgi birimi. Aynı anda hem 0 hem 1 olabilir.
### Reversibilite (Döndürülebilirlik) =
Kuantum kapılarının tersine çevrilebilir olması, yani işlemin geri alınabilmesi.
### Reel Kısım = 
Karmaşık sayının gerçek kısmı.
### Rotasyon Kapıları (Rx, Ry, Rz) = 
Qubit’i Bloch küresinde x, y veya z eksenleri etrafında döndürür.
### Sabit Fonksiyon = 
Tüm girdiler için aynı çıktıyı veren fonksiyon.
### Sanal Kısım = 
Karmaşık sayının katsayılı kısmı.
### Sayma Qubitleri (Counting Qubits) = 
Faz bilgisini kaydetmek için kullanılan yardımcı qubitler.
### Schrödinger Denklemi = 
Kuantum mekaniğinde parçacığın dalga fonksiyonunu belirleyen temel diferansiyel denklem.
### Shor Algoritması = 
Büyük sayıların asal çarpanlarını bulan ünlü kuantum algoritması.
### Simülasyon / Simülatör = 
Gerçek kuantum cihazı olmadan bilgisayarda kuantum devrelerini çalıştırma.
### Spin = 
Parçacıkların sahip olduğu kuantum mekaniğine özgü içsel açısal momentum.
### Statevector = 
Qubit’in matematiksel durumunu gösteren vektör.
### Süper Yoğun Kodlama (Superdense Coding) = 
Dolanıklık kullanarak, tek bir qubit ile iki klasik bitlik bilgi gönderme yöntemi.
### Süperpozisyon = 
Qubit’in aynı anda hem 0 hem 1 olabilmesi.
### SWAP Kapısı = 
İki qubit’in yerini değiştiren kapı.
### T-Gate (T Kapısı) = 
Faz kapısının özel bir versiyonu. Fazı π/4 kadar kaydırır.
### Tek Qubit Kapıları = 
Bir qubit üzerinde etki eden kuantum kapıları.
### Teleportasyon = 
Qubit’in durumunu başka bir yere aktarma yöntemi. (Burada karşılaştırma için geçiyor.)
### Tensor Çarpımı = 
İki vektör uzayını birleştiren işlem; çok qubitli sistemleri ifade etmek için kullanılır.
### Ters Fourier Dönüşümü (Inverse QFT) = 
Fourier uzayındaki bilgiyi tekrar normal hesaplama uzayına çeviren dönüşüm.
### Toffoli (CCNOT) Kapısı = 
İki kontrol qubit’i aynı anda 1 olduğunda hedef qubit’i çeviren kapı.
### Transpile = 
Yazılan kuantum devresini, seçilen cihaz ya da simülatöre uygun hale getirme işlemi.
### Üniter Matris = 
Tersi kendisinin eşleniği olan matris. Qubit işlemlerinde olasılıkların korunmasını sağlar.
### Üniter Operatör = 
Sistemin zamanla evrimini belirleyen, olasılıkların korunmasını sağlayan operatör.
### Varyans = 
Olası sonuçların ortalamadan ne kadar uzaklaştığını gösteren ölçü.
### Vektör = 
Yönü ve büyüklüğü olan matematiksel nesne; kuantumda parçacık durumlarını temsil eder.
### VQE (Variational Quantum Eigensolver) = 
Kuantum bilgisayarlarda enerji hesaplamak için kullanılan algoritma.
### XZ Operatörleri = 
Pauli-X ve Pauli-Z kapılarının birleşimi; süper yoğun kodlama protokolünde kullanılır.
### XOR Mantığı = 
İki girdiden yalnızca biri 1 olduğunda çıkışı 1 yapan klasik mantık işlemi; CNOT kapısının klasik karşılığıdır.
